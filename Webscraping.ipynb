```python

import time
import csv
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.expected_conditions import presence_of_element_located
import os
import urllib.request
from urllib.error import HTTPError

dir_path = "D:\\Jen\\Documents\\Dissertation\\Bee_Data_Groups\\"
csv_path = dir_path + "Bee_Data.csv"
new_csv_path = dir_path + "Bee_Data_with_metadata.csv"
grouped_csv_path = dir_path + "Bee_Data_Grouped.csv"

def download_image(url, num):
    if url is None:
        no_image = True
    else:
        print("[INFO] downloading {}".format(url))
        name = str(url.split('/')[-1])
        no_image = False
        try:
            urllib.request.urlretrieve(url, dir_path + file_name + str(num) + ".jpg")
        except:
            no_image = True
    return no_image

with webdriver.Firefox() as driver:
    wait = WebDriverWait(driver, 10)
    driver.get("https://www.brc.ac.uk/irecord")
    driver.find_element(By.ID, "edit-name").send_keys("jmollett1@sheffield.ac.uk" + Keys.TAB)
    driver.find_element(By.ID, "edit-pass").send_keys(password + Keys.ENTER)
    time.sleep(15)
    with open(csv_path, 'r', newline='') as f:
        with open(new_csv_path, 'a', newline='') as f_new:
            beereader = csv.reader(f)
            beewriter = csv.writer(f_new)
            for row in beereader:
                file_name = row[8]
                driver.get("https://www.brc.ac.uk/irecord/record-details?occurrence_id=" + row[0])
                time.sleep(2)
                #print(driver.find_element_by_xpath("//*[@id='detail-panel-recorddetails']/div[1]/dd[8]").text)
                labels = ["EUNIS Habitat", "Stage", "Sex", "Quantity", "Certainty", "Record status"]
                details = {"Record status": "", "EUNIS Habitat": "", "Stage": "", "Certainty": "", "Sex": "", "Quantity": ""}
                found = False
                categories = driver.find_elements(By.CLASS_NAME, "record-details-fields")
                try:
                    elements = categories[0].find_elements_by_xpath(".//*") + categories[1].find_elements_by_xpath(".//*")
                except:
                    for label in labels:
                        row.append(details[label])
                    beewriter.writerow(row)
                    continue
                for element in elements:
                    if found:
                        details[found] = element.text
                        found = False
                    if element.text in labels:
                        found = element.text
                for label in labels:
                    row.append(details[label])
                beewriter.writerow(row)
                images = driver.find_elements(By.CLASS_NAME, "fancybox")
                if len(images) > 1:
                    with open(grouped_csv_path, 'a', newline='') as f_group:
                        beegroupwriter = csv.writer(f_group)
                        beegroupwriter.writerow(row)
                        image_num = 2
                        for image in images[1:]:
                            postfix = "-" + str(image_num)
                            no_image = download_image(image.get_attribute("href"), postfix)
                            if not no_image:
                                image_num += 1
                                new_file_name = file_name + postfix
                                row[8] = new_file_name
                                beegroupwriter.writerow(row)
```
